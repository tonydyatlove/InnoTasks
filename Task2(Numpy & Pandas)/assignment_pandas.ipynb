## Проведите анализ данных при помощи Pandas выполнив поставленные задачи.
#### 

```python
import pandas as pd
```

```python
# загружаем датасет
data = pd.read_csv("adult.data.csv")
data.head()
```

**1. Посчитайте, сколько мужчин и женщин (признак *sex*) представлено в этом датасете**

```python
male_count = data.loc[data['sex'] == "Male"].count()['sex']
print(f'Количество мужчин: {male_count}')
female_count = data.loc[data['sex'] == "Female"].count()['sex']
print(f'Количество женщин: {female_count}')
```

**2. Каков средний возраст мужчин (признак *age*) по всему датасету?**

```python
mean_age_male = data.loc[data['sex'] == 'Male', 'age'].mean().round(2)
print(f'Средний возраст мужчин: {mean_age_male}')
```

**3. Какова доля граждан Соединенных Штатов (признак *native-country*)?**

```python
us_native = data['native-country'].value_counts(normalize=True)['United-States'].round(3)
print(f'Доля граждан Соединённых Штатов: {us_native}')
```

**4-5. Рассчитайте среднее значение и среднеквадратичное отклонение возраста тех, кто получает более 50K в год (признак *salary*) и тех, кто получает менее 50K в год**

```python
mean_age_higher50k = data.loc[data['salary'] == '>50K', 'age'].mean().round(2)
print(f'Средний возраст людей с доходом >50K в год: {mean_age_higher50k}')

std_age_higher50k = round(data.loc[data['salary'] == '>50K', 'age'].std(), 2)
print(f'Стандартное отклонение возраста людей с доходом >50K в год: {std_age_higher50k}')

mean_age_lower50k = data.loc[data['salary'] == '<=50K', 'age'].mean().round(2)
print(f'Средний возраст людей с доходом <=50K в год: {mean_age_lower50k}')

std_age_lower50k = round(data.loc[data['salary'] == '<=50K', 'age'].std(), 2)
print(f'Стандартное отклонение возраста людей с доходом <=50K в год: {std_age_lower50k}')
```

**6. Правда ли, что люди, которые получают больше 50k, имеют минимум высшее образование? (признак *education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters* или *Doctorate*)**

```python
education_list = ['Bachelors', 'Prof-school', 'Assoc-acdm', 'Assoc-voc', 'Masters', 'Doctorate']
education_higher50k = round(data.loc[(data['education'].isin(education_list)) & 
                                (data['salary'] == '>50K')].count() / data.loc[data['salary'] == '>50K'].count(), 3)[0]
print(f'Доля людей, получающих >50K, которые имеют высшее образование: {education_higher50k}')
# Правда отчасти, так как не все люди, получающие больше 50k, имеют высшее образование.
# Однако, большинство(58%) таких людей имеют высшее образование
```

**7. Выведите статистику возраста для каждой расы (признак *race*) и каждого пола. Используйте *groupby* и *describe*. Найдите таким образом максимальный возраст мужчин расы *Asian-Pac-Islander*.**

```python
gender_race_stats = data.groupby(['race', 'sex'])['age'].describe().reset_index()
gender_race_stats
```

```python
max_age_API_race = gender_race_stats.loc[(gender_race_stats['sex'] == 'Male') &
                                         (gender_race_stats['race'] == 'Asian-Pac-Islander')]['max'].values[0]
print(f'Максимальный возраст мужчин расы Asian-Pac-Islander: {max_age_API_race}')
```

**8. Среди кого больше доля зарабатывающих много (>50K): среди женатых или холостых мужчин (признак *marital-status*)? Женатыми считаем тех, у кого *marital-status* начинается с *Married* (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), остальных считаем холостыми.**

```python
married_list = ['Married-civ-spouse', 'Married-spouse-absent', 'Married-AF-spouse']

married_higher50k = round(data.loc[(data['marital-status'].isin(married_list)) & 
                                (data['salary'] == '>50K') &
                                     (data['sex'] == 'Male')].count() / data.loc[(data['salary'] == '>50K') &
                                                                                      (data['sex'] == 'Male')].count(), 3)[0]
print(f'Доля женатых мужчин, среди зарабатывающих >50K: {married_higher50k}')
print(f'Доля неженатых мужчин, среди зарабатывающих >50K: {(1 - married_higher50k).round(3)}')
# Доля женатых почти в 9 раз больше!
```

**9. Какое максимальное число часов человек работает в неделю (признак *hours-per-week*)? Сколько людей работают такое количество часов и каков среди них процент зарабатывающих много?**

```python
max_hours_per_week = data['hours-per-week'].max()
print(f'Максимальное количество рабочих часов в неделю: {max_hours_per_week}')

amount = data.loc[data['hours-per-week'] == max_hours_per_week].count()[0]
print(f'Количество людей, работающих столько часов: {amount}')

big_salary_percentage = data.loc[data['hours-per-week'] == max_hours_per_week]['salary'].value_counts(normalize=True)['>50K'].round(3)
print(f'Доля тех, кто зарабатывает >50K: {big_salary_percentage}')
```

**10. Посчитайте среднее время работы (*hours-per-week*) зарабатывающих мало и много (*salary*) для каждой страны (*native-country*).**

**11.Сгруппируйте людей по возрастным группам *young*, *adult*, *retiree*, где:**
* *young* соответствует 16-35 лет
* *adult* - 35-70 лет
* *retiree* - 70-100 лет

**Проставьте название соответсвтуещей группы для каждого человека в новой колонке AgeGroup**

```python
bins = [16, 35, 70, 100]
labels = ['young', 'adult', 'retiree']

data['AgeGroup'] = pd.cut(data['age'], bins=bins, labels=labels, right=False)

data.head(10)
```

**12-13. Определите количество зарабатывающих >50K в каждой из возрастных групп (колонка AgeGroup), а также выведите название возрастной группы, в которой чаще зарабатывают больше 50К (>50K)**

```python
age_group_salary = data.loc[data['salary'] == '>50K'].groupby('AgeGroup')['salary'].count().to_frame().reset_index()
print(f'Количество зарабатывающих >50K в каждой из возрастных групп:')
print(age_group_salary)
max_value = age_group_salary['salary'].max()
name = age_group_salary.loc[age_group_salary['salary'] == max_value]['AgeGroup'].values[0]
print(f'Название самой многочисленной группы с зарплатой >50K: {name}')
```

**14. Сгруппируйте людей по типу занятости (колонка occupation) и определите количество людей в каждой группе. После чего напишите функциюю фильтрации filter_func, которая будет возвращать только те группы, в которых средний возраст (колонка age) не больше 40 и в которых все работники отрабатывают более 5 часов в неделю (колонка hours-per-week)**

```python
occupation_distribution = data['occupation'].value_counts()
print(f'Количество людей в каждой группе: ')
print(occupation_distribution)
```

```python
def filter_func(df):
    return (df['age'].mean() <= 40) & (df['hours-per-week'].min() > 5)

grouped = data.groupby('occupation')

filtered_data = grouped.filter(filter_func)['occupation'].values[0]
print(f'Группы, в которых средний возраст не больше 40 и все работники отрабатывают более 5 часов в неделю: {filtered_data}')
```
